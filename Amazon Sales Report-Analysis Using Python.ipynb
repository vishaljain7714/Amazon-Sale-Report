## Amazon Sales Analysis-An Analysis Using Python

```python
# Let us import the necessary library(ies)
import pandas as pd

# Reading the Dataset:
df=pd.read_excel('asr.xlsx')
df
```

#### Sales Overview

```python
# Here we will load the dataframe (assuming it's already loaded as 'df')
# Example: df = pd.read_csv('my_data.csv')

# Calculate overall sales performance
total_sales = df['Amount'].sum()

#Then we will  Analyze trends over time (assuming 'Date' column is datetime)
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Amount'].sum()

#Finally Print or return results
print(f"Total Sales: ${total_sales}")
print("Monthly Sales Trends:")
print(monthly_sales)
```

### Product Analysis

```python
# First we will Analyze product categories, sizes, and quantities sold
category_counts = df['Category'].value_counts()
size_counts = df['Size'].value_counts()

#Then Print or return results
print("Product Categories:")
print(category_counts)
print("\nProduct Sizes:")
print(size_counts)
```

### Fullfillment Analysis

```python
# First we will Investigate fulfillment methods and their effectiveness
fulfillment_counts = df['Fulfilment'].value_counts()

#Then we will Print or return results
print("Fulfillment Methods:")
print(fulfillment_counts)
```

### Customer Segmentation

```python
# First we will Segment customers based on buying behavior (e.g., amount spent)
bins = [0, 100, 500, 1000, float('inf')]
labels = ['Low', 'Medium', 'High', 'VIP']
df['Customer Segment'] = pd.cut(df['Amount'], bins=bins, labels=labels)

# Then Analyze segments
segment_counts = df['Customer Segment'].value_counts()

# Finallywe will Print or return results
print("Customer Segments:")
print(segment_counts)
```

### Geographical Analysis

```python
# Let us Explore geographical distribution of sales
state_sales = df.groupby('ship-state')['Amount'].sum()
city_sales = df.groupby('ship-city')['Amount'].sum()

#Then Print or return results
print("State-wise Sales:")
print(state_sales)
print("\nCity-wise Sales:")
print(city_sales)

```

```python
# First we will Calculate state-wise and city-wise sales
state_sales = df.groupby('ship-state')['Amount'].sum()
city_sales = df.groupby('ship-city')['Amount'].sum()

#Then we will Sort state-wise sales
sorted_state_sales = state_sales.sort_values(ascending=False)
top_5_states = sorted_state_sales.head(5)
bottom_5_states = sorted_state_sales.tail(5)

# Similarly we will be Sorting city-wise sales
sorted_city_sales = city_sales.sort_values(ascending=False)
top_5_cities = sorted_city_sales.head(5)
bottom_5_cities = sorted_city_sales.tail(10)

#Finally we will Print or return results
print("Top 5 States by Sales:")
print(top_5_states)
print("\nBottom 5 States by Sales:")
print(bottom_5_states)

print("\nTop 5 Cities by Sales:")
print(top_5_cities)
print("\nBottom 5 Cities by Sales:")
print(bottom_5_cities)
```
